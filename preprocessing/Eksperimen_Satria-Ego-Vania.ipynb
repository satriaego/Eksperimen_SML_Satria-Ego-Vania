#### Analisis & Penanganan Missing Values:
1. Kolom Cabin: Dihapus (drop) dari dataset karena persentase data yang hilang mencapai >78% (327 dari 418 baris). Mengisi data dengan tebakan pada skala sekecil ini dapat memanipulasi keaslian informasi.
2. Kolom Age & Fare: Mengandung data kosong dalam jumlah minor, sehingga diimputasi menggunakan nilai Median agar tidak sensitif terhadap outlier.

```python
df = df.drop(columns=['Cabin'], errors='ignore') 
df['Age'] = df['Age'].fillna(df['Age'].median())  
df['Fare'] = df['Fare'].fillna(df['Fare'].median()) 

```

```python
print("Jumlah data kosong:")
print(df.isnull().sum())
print(f"\nJumlah data duplikat: {df.duplicated().sum()}")
```

```python
# Kolom Nama, ID, dan Nomor Tiket tidak mempengaruhi keselamatan
kolom_tidak_butuh = ['PassengerId', 'Name', 'Ticket']
df_selected = df.drop(columns=kolom_tidak_butuh, errors='ignore')

# Mengubah kolom 'Sex' dan 'Embarked' menjadi angka biner (0 dan 1) agar bisa dibaca model ML
df_ready = pd.get_dummies(df_selected, columns=['Sex', 'Embarked'], drop_first=True)


print("✅ Preprocessing Selesai!")
print(f"Ukuran dataset saat ini: {df_ready.shape[0]} baris dan {df_ready.shape[1]} kolom.\n")
df_ready.head()
```
